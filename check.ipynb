# Question 1: Implementation of single perceptron 

```python

# METHOD 3: BUILT-IN PERCEPTRON

# Uses sklearn library


from sklearn.linear_model import Perceptron

# Input data (3 inputs per sample)
X = [
    [4, 3, 2],
    [2, 1, 1],
    [5, 2, 3],
    [1, 1, 0]
]

# Output labels
y = [1, 0, 1, 0]

# Create perceptron model
model = Perceptron()

# Train the model
model.fit(X, y)

# Test input
test_input = [[4, 3, 2]]

# Predict output
prediction = model.predict(test_input)

# Print result
print("Built-in Perceptron Output:", prediction[0])
```

# Question 2: Real life example

University Admission Decision
Problem statement:
A university wants to decide whether to give admission or not based on 3 factors
1. Entry test score (out of 100)
2. Intermediate percentage
3. Interview score (out of 10)

```python


# Step function
def step_function(z):
    if z >= 0:
        return 1
    else:
        return 0


# Perceptron function (3 inputs)
def perceptron(x1, x2, x3, w1, w2, w3, bias):
    
    # Calculate weighted sum
    z = (x1 * w1) + (x2 * w2) + (x3 * w3) + bias
    
    print("Weighted Sum (z):", z)
    
    # Return decision
    return step_function(z)


# Call perceptron
decision = perceptron(70, 75, 8, 0.05, 0.04, 0.3, -6)

print("Admission Decision:", decision)
```

```python
import math

def sigmoid(z):
    # The formula: 1 / (1 + e^-z)
    return 1 / (1 + math.exp(-z))

# Inputs
x1, x2 = 1, 2

# Layer 1 parameters
w1, w2 = 0.1, 0.2
b1 = -2

# Layer 2 parameters
w3 = 0.3
b2 = 0.1

def forward_pass_with_sigmoid(x1, x2):
    # Step 1: First node calculation
    z1 = (x1 * w1) + (x2 * w2) + b1  # Linear sum: 0.1 + 0.4 - 2 = -1.5
    h1 = sigmoid(z1)                # Apply Sigmoid: 1 / (1 + e^1.5)
    print(f"Node 1 Linear Sum (z1): {z1}")
    print(f"Node 1 after Sigmoid (h1): {round(h1, 4)}")

    # Step 2: Final output node calculation
    z2 = (h1 * w3) + b2             # Linear sum: (h1 * 0.3) + 0.1
    y_hat = sigmoid(z2)             # Apply Sigmoid again
    
    return y_hat

# Execute
result = forward_pass_with_sigmoid(x1, x2)
print(f"---")
print(f"Final output (y_hat) with Sigmoid: {round(result, 4)}")
```

```python
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Input values
x1, x2, x3 = 1, 0, 1

# Weights (Input to Hidden)
w14, w15 = 0.2, -0.3
w24, w25 = 0.4, 0.1
w34, w35 = -0.5, 0.2

# Weights (Hidden to Output)
w46, w56 = -0.3, -0.2

# Biases
b4, b5 = -0.4, 0.2
b6 = 0.1

def run_mlp():
    # Calculation for Layer 1 (Hidden)
    sum_h4 = (x1 * w14) + (x2 * w24) + (x3 * w34) + b4
    out_h4 = sigmoid(sum_h4)
    
    sum_h5 = (x1 * w15) + (x2 * w25) + (x3 * w35) + b5
    out_h5 = sigmoid(sum_h5)
    
    print(f"h4: Sum = {sum_h4}, Output = {out_h4}")
    print(f"h5: Sum = {sum_h5}, Output = {out_h5}")

    # Calculation for Layer 2 (Output)
    sum_o6 = (out_h4 * w46) + (out_h5 * w56) + b6
    out_o6 = sigmoid(sum_o6)
    
    return out_o6

# Execution
final_output = run_mlp()
print("-" * 40)
print(f"Final Output O6: {final_output}")
```

```python
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# --- Inputs ---
x1, x2 = 1, 2

# --- Layer 1 (Hidden) ---
w1, w2 = 0.1, 0.2  # Weights for a1
w3, w4 = 0.3, 0.4  # Weights for a2
b1, b2 = -1, -2    # Biases for a1, a2

# --- Layer 2 (Hidden) ---
w5, w6 = 0.5, 0.6  # Weights for a3
w7, w8 = 0.7, 0.8  # Weights for a4
b3, b4 = -3, -4    # Biases for a3, a4

# --- Output Layer ---
w9, w10 = 0.9, 0.8 # Weights for Output
b5 = -5            # Bias for Output

def run_part_3():
    # 1. Hidden Layer 1 Calculations
    z_a1 = (x1 * w1) + (x2 * w2) + b1  # (1*0.1 + 2*0.2) - 1 = -0.5
    out_a1 = sigmoid(z_a1)
    
    z_a2 = (x1 * w3) + (x2 * w4) + b2  # (1*0.3 + 2*0.4) - 2 = -0.9
    out_a2 = sigmoid(z_a2)
    
    print(f"Layer 1 -> a1: {round(out_a1, 4)}, a2: {round(out_a2, 4)}")

    # 2. Hidden Layer 2 Calculations 
    # (These nodes receive inputs from both a1 and a2)
    z_a3 = (out_a1 * w5) + (out_a2 * w6) + b3
    out_a3 = sigmoid(z_a3)
    
    z_a4 = (out_a1 * w7) + (out_a2 * w8) + b4
    out_a4 = sigmoid(z_a4)
    
    print(f"Layer 2 -> a3: {round(out_a3, 4)}, a4: {round(out_a4, 4)}")

    # 3. Output Layer Calculation
    z_yhat = (out_a3 * w9) + (out_a4 * w10) + b5
    y_hat = sigmoid(z_yhat)
    
    return y_hat

# Execute
final_result = run_part_3()
print("-" * 40)
print(f"Final Output (y_hat): {final_result}")
```

```python
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Inputs (Scaled down)
x1, x2 = 0.6, 0.7

# Weights & Biases: Layer 1
w11, w12 = 0.92, 0.32
w21, w22 = 0.45, 0.65
b1, b2 = 0.11, 0.12

# Weights & Biases: Layer 2
w13, w14 = 0.11, 0.14
w23, w24 = 0.23, 0.24
b3, b4 = 0.13, 0.14

# Weights & Biases: Layer 3
w35, w36 = 0.35, 0.36
w45, w46 = 0.45, 0.46
b5, b6 = 0.15, 0.16

# Weights & Biases: Output Layer
w5, w6 = 0.52, 0.62
b7 = 0.17

def run_functional_mlp():
    # --- Layer 1 ---
    a1 = sigmoid((x1 * w11) + (x2 * w21) + b1)
    a2 = sigmoid((x1 * w12) + (x2 * w22) + b2)
    print(f"Layer 1 Outputs: a1={round(a1, 4)}, a2={round(a2, 4)}")

    # --- Layer 2 ---
    a3 = sigmoid((a1 * w13) + (a2 * w23) + b3)
    a4 = sigmoid((a1 * w14) + (a2 * w24) + b4)
    print(f"Layer 2 Outputs: a3={round(a3, 4)}, a4={round(a4, 4)}")

    # --- Layer 3 ---
    a5 = sigmoid((a3 * w35) + (a4 * w45) + b5)
    a6 = sigmoid((a3 * w36) + (a4 * w46) + b6)
    print(f"Layer 3 Outputs: a5={round(a5, 4)}, a6={round(a6, 4)}")

    # --- Output ---
    o7 = sigmoid((a5 * w5) + (a6 * w6) + b7)
    return o7

final_val = run_functional_mlp()
print("-" * 30)
print(f"Final Model Output: {round(final_val, 4)}")
```
